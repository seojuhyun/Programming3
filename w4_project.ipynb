```python
'''
#1 과제 1 -> movie 데이터의 크기를 줄여봅시다. 

movie.info()

movie데이터가 많기때문에 

movie1= movie.iloc[: ,:11].fillna(0)

movie1.info()   --->이것을 줄이면 됨!! ex) category..
'''

import pandas as pd
import numpy as np

movie= pd.read_csv("data/movie.csv")

movie1= movie.iloc[:,:11].fillna(0)
movie1.info()
 
```

```python
movie1.memory_usage(deep=True)
```

```python
movie1.describe().T
```

```python
movie1.describe(include='O').T
```

```python
np.finfo('float64')
```

```python
movie1['num_critic_for_reviews']= movie1['num_critic_for_reviews'].astype(np.float32)
movie1['duration']= movie1['duration'].astype(np.float32)
movie1['director_facebook_likes']= movie1['director_facebook_likes'].astype(np.float32)
movie1['actor_3_facebook_likes']= movie1['actor_3_facebook_likes'].astype(np.float32)
movie1['actor_1_facebook_likes']= movie1['actor_1_facebook_likes'].astype(np.float32)
movie1['gross']= movie1['gross'].astype(np.float32)

movie1.memory_usage(deep=True)
```

```python
movie1['color']= movie1['color'].astype('category')
movie1['genres']= movie1['genres'].astype('category')

movie1.memory_usage(deep=True)
```

```python
'''
#2 과제 2 -> college 에서 UGDS 의 최대값을 구하는데 도시별로 구하시오.

hint) groupby, nlargest, dropduplicates..

college.head()

#INSTNM, CITY, UGDS 
'''

college= pd.read_csv('data/college.csv')
college[['INSTNM', 'CITY', 'UGDS']].head()


college[['INSTNM', 'CITY', 'UGDS']].groupby('CITY', as_index=False).apply(lambda df: df.sort_values(
    'UGDS', ascending= False).head(1)).droplevel(0).sort_values('CITY', ascending=False)
          
```

```python
college[['INSTNM', 'CITY', 'UGDS']].sort_values(['CITY','UGDS'], ascending=False).drop_duplicates(subset='CITY')
```
