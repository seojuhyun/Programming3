```python
# 과제2

movie= pd.read_csv('data/movie.csv', nrows=1000)
movie.info()
```

```python
movie.describe().T
```

```python
#object인 것들 중 category로 바꿀 수 있는것이 무엇인지 확인한다.(unique가 10이하인 것을 category로 만들어준다.) 

movie.describe(include='O').T
```

```python
movie.columns
```

```python
#사용할 columns 중 이름이 긴 것을 보기 좋게 바꿔준다. 

cols_map={'num_critic_for_reviews':'critic_reviews',
          'cast_total_facebook_likes':'total_likes',
          'movie_facebook_likes':'movie_likes'}
          
movies=movie.rename(columns=cols_map)
movies.to_csv('movies.csv')

#메모리 용량을 줄이기 위해서 사용할 columns만 뽑아서 정리한다. 

cols=['color','director_name','critic_reviews','duration','genres','total_likes', 'movie_title','language','title_year','movie_likes']

#메모리 용량을 줄이기 위해서 category로 바꿀 수 있는 object는 category로 바꿔준다. 
#그 외에도 float64는 float16으로, int64는 int32로 바꿔준다. 

movies_dty= pd.read_csv('movies.csv', nrows=1000,
                        dtype={
                        'color': 'category',
                        'language': 'category',
                        'critic_reviews': np.float16,
                        'duration': np.float16,
                        'total_likes': np.int32, 
                        'title_year': np.float16,
                        'movie_likes': np.int32
                        },
                        usecols= cols)

movies_dty.info()
```

다음과 같은 과정을 통해 기존에 사용했던 메모리는 5배 넘게 줄어든 것을 확인 할 수 있다. 

기존 사용 메모리= 218.9
줄어든 메모리= 39.7

```python
round(218.9/39.7, 2)
```
